# 05 — word2vec (CBOW vs. Skip‑gram) — Conceptual Playground

This notebook demonstrates the shapes and cosine similarities of random embeddings; replace with a real training setup if running locally with gensim or PyTorch.

```
import numpy as np
np.random.seed(0)
V = ["boy","girl","king","queen","apple","mango"]
E = {w: np.random.randn(6) for w in V}
from numpy.linalg import norm
cos = lambda u,v: float(np.dot(u,v)/(norm(u)*norm(v)))
for a in V:
    for b in V:
        if a<b:
            print(f"cos({a},{b})=", round(cos(E[a],E[b]),3))
```
